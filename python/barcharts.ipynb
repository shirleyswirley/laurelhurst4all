```python
import matplotlib.pyplot as plt
import numpy as np
```

## Draw final figure - poster 3

```python
labels_energy_uses = ['Embodied', 'Operational', 'Transportation', 'Total']
passive_sf_home = 4832*np.array([0.44, 0.31, 0.25, 1])
std_sf_home = 4854*np.array([0.38, 0.37, 0.25, 1])
std_apt = 2830*np.array([0.31, 0.44, 0.25, 1])
```

```python
fig,ax = plt.subplots(1, 1, figsize = (6,6))

x = np.arange(len(passive_sf_home))
width = 0.35

#rects1 = ax.barh(x+width, passive_sf_home, width, label='Passive single family home')
#rects2 = ax.barh(x, std_sf_home, width, label='Single family home')
#rects3 = ax.barh(x-width, std_apt, width, label='Apartment')
rects1 = ax.barh(x+width/2, std_sf_home, width, label='Single family home')
rects2 = ax.barh(x-width/2, std_apt, width, label='Apartment')

for i, v in enumerate(std_sf_home):
    ax.text(v + 15, i + 0.1, format(round(v, -1).astype(int), ','), color='black', fontsize=12, fontweight='bold')
    
for i, v in enumerate(std_apt):
    ax.text(v + 15, i - 0.28, format(round(v, -1).astype(int), ','), color='black', fontsize=12, fontweight='bold')

ax.set_yticks(x)
ax.set_yticklabels(labels_energy_uses, fontsize=16)
ax.tick_params(axis='x', labelsize=14)
ax.set_xlabel('Energy use over 50 years [GJ/capita]', fontsize=16)
ax.spines["right"].set_visible(False)
#ax.spines["bottom"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.tick_params(left=False)
ax.legend(frameon=False, loc='center right')

fig.savefig('/opt/laurelhurst4all/figures/poster3.pdf', dpi=400, bbox_inches='tight', pad_inches = 0)
fig.savefig('/opt/laurelhurst4all/figures/poster3.png', dpi=400, bbox_inches='tight', pad_inches = 0)
```

## Draw final figure - poster 4

```python
labels_race = ['White ', 'Black ', 'Latinx ']
wealth = [456000, 23000, 90000]
home_ownership = [50.9, 25.8, 26.5]

labels_or = ['Own', 'Rent']
med_net_worth = [898000, 36000]
```

```python
fig,axes = plt.subplots(1, 2, figsize = (10,3))

x = np.arange(len(labels_race))  # the label locations
width = 0.7  # the width of the bars

ax = axes[0]
rects = ax.barh(x, wealth, width)
ax.set_title('Median household net worth', fontsize=25)
for i, v in enumerate(wealth):
    ax.text(v + 9000, i - 0.11, '$' + format(v, ','), color='black', fontsize=16, fontweight='bold')

ax = axes[1]
rects = ax.barh(x, home_ownership, width)
ax.set_title('Home ownership', fontsize=25)
for i, v in enumerate(home_ownership):
    ax.text(v + 1, i - 0.11, str(v) + '%', color='black', fontsize=16, fontweight='bold')
        
for ax in axes:
    ax.set_yticks(x)
    ax.set_yticklabels(labels_race, fontsize=14)
    ax.spines["right"].set_visible(False)
    ax.spines["bottom"].set_visible(False)
    ax.spines["top"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.tick_params(axis=u'both', which=u'both',length=0)
    ax.axes.get_xaxis().set_visible(False)

fig.tight_layout()

fig.savefig('/opt/laurelhurst4all/figures/poster4.pdf', dpi=400, bbox_inches='tight', pad_inches = 0)
fig.savefig('/opt/laurelhurst4all/figures/poster4.png', dpi=400, bbox_inches='tight', pad_inches = 0)
```
